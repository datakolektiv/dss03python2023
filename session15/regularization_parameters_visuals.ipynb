```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
from sklearn.datasets import make_regression

from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
X, y, coef = make_regression(noise=30, n_features=1, n_informative=1, n_targets=1, n_samples=100, coef=True)
```

```python
X.shape, y.shape, coef.shape
```

```python
lr = LinearRegression();

lr.fit(X, y)
```

```python
y_preds = lr.predict(X)
```

```python
sns.scatterplot(x=X[:, 0], y=y, linewidth=0, label='observations')
sns.scatterplot(x=X[:, 0], y=y_preds, c='r', linewidth=0, label='predictions');
```

```python
sns.scatterplot(x=X[:, 0], y=y-y_preds)
plt.axhline(y=0, c='r');
```
